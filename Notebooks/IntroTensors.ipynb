### 2. Testing a 1D Tensor/Vector

- [1,2,3,4] its a 1d tensor and its ndim is 1
- it can be a vector as well and 1d array or just an array
- its a 1d tensor but its n of elements decide what D vector it is e.g. [1,2,3] then its a 1D tensor but its a 3D Vector.

```python
arr = np.array([1,2,3,4])
```
